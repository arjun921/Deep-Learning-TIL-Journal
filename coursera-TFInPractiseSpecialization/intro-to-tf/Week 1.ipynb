```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
model = keras.Sequential(
    [
        keras.layers.Dense(units=1, input_shape=[1])
    ]
)
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
x = np.array([
    -1.0,
    0.0,
    1.0,
    2.0,
    3.0,
    4.0
], dtype=float)
y = np.array([
    -3.0,
    -1.0,
    1.0,
    3.0,
    5.0,
    7.0
], dtype=float)
```

```python
# epochs = model will go through training loop 500 times
# 1 loop = make a guess, measure how good or bad the guess is using loss function, use optimizer and data to make another guess
model.fit(x,y,epochs=500)
```

```python
model.predict([10.0])
```
